```python
import pandas as pd
import pandas_ta as ta
import numpy as np
import yfinance as yf
import vectorbt as vbt
import datetime
```

```python
import mplfinance as mpf
```

## Download Historical Price Data

```python
price_df = yf.Ticker('BTC-USD').history(start=datetime.datetime(2021, 1, 1),
                                        end=datetime.datetime(2022, 4, 30),
                                        interval = '1D')
```

```python
price_df = price_df[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
price_df 
```

## Signal Calculation

```python
ema_period = 15
```

```python
df = price_df.copy()
```

### Calculate DEMA

```python
df.ta.ema(ema_period, append=True)
```

```python
df.ta.sma(ema_period, append=True)
```

```python
df['2EMA_15'] = ta.ema(df['EMA_15'], ema_period)
```

```python
df['DEMA'] = (2 * df['EMA_15']) - df['2EMA_15']
```

```python
df
```

### Calculate MACD

```python
df.ta.macd(append=True)
```

### Calculate BB Bands

```python
df.ta.bbands(20, 2.0, mamode='ema', append=True)
```

```python
df.ta.rsi(append=True)
```

```python
df.ta.adx(append=True)
```

### Calculate Signal

```python
# df['signal_buy'] = df.apply(lambda _ : ((_['Close'] > _['DEMA']) or (_['MACD_12_26_9'] > _['MACDs_12_26_9']))
#                             and _['BBB_20_2.0'] >= 15,
#                             axis=1)

df['signal_buy'] = df.apply(lambda _ : ((_['Close'] > _['EMA_15']) or (_['MACDh_12_26_9'] > 0))
                            and _['BBB_20_2.0'] >= 15,
                            axis=1)

# df['signal_buy'] = df.apply(lambda _ : _['DEMA'] > _['EMA_15'],
#                             axis=1)


# df['signal_buy'] = df.apply(lambda _: (_['DEMA'] > _['EMA_15']) and (_['BBB_20_2.0'] >= 15), 
                            
#                             axis=1)
```

```python
# df['signal_sell'] = df.apply(lambda _ : (_['Close'] < _['DEMA']) or (_['MACD_12_26_9'] < _['MACDs_12_26_9']), 
#                             axis=1)

df['signal_sell'] = df.apply(lambda _ : (_['Close'] < _['EMA_15']) or (_['MACDh_12_26_9'] < 0), 
                            axis=1)

# df['signal_sell'] = df.apply(lambda _ : _['DEMA'] < _['EMA_15'], 
#                             axis=1)

# df['signal_sell'] = df.apply(lambda _: _['DEMA'] < _['EMA_15'], axis=1)
```

## Prepare signaling DataFrame for VBT

```python
buy_signal_df = df.ta.tsignals(df['signal_buy'], asbool=True, append=True)
```

```python
sell_signal_df = df.ta.tsignals(df['signal_sell'], asbool=True, append=True)
```

```python
buy_signal_df['TS_Entries'] = buy_signal_df['TS_Entries'].shift(1).fillna(False)
```

```python
sell_signal_df['TS_Entries'] = sell_signal_df['TS_Entries'].shift(1).fillna(False)
```

## Backtesting

```python
port = vbt.Portfolio.from_signals(df['Open'],
                                 entries=buy_signal_df['TS_Entries'], # Long Entry
                                 exits=sell_signal_df['TS_Entries'], # Long Exit
                                 freq='D',
                                 init_cash=100_000,
                                 sl_stop=0.05,
                                 fees=0.0025,
                                 slippage=0.0025)
```

```python
port.stats()
```

```python
port.plot().show()
```

```python
sport = vbt.Portfolio.from_signals(df['Open'],
                                 short_entries=sell_signal_df['TS_Entries'], # Long Entry
                                 short_exits=buy_signal_df['TS_Entries'], # Long Exit
                                 freq='D',
                                 init_cash=100_000,
                                 sl_stop=0.08,
                                 fees=0.0025,
                                 slippage=0.0025)
```

```python
sport.stats()
```
